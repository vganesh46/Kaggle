```python
import pandas as pd
from nltk.tokenize import word_tokenize
import re
from nltk.corpus import stopwords
```

```python
train = pd.read_csv("/Users/ganeshshankar/Desktop/work_mac/Personal_Related/Learning/Datascience/Kaggle/Quora/train.csv")
test = pd.read_csv("/Users/ganeshshankar/Desktop/work_mac/Personal_Related/Learning/Datascience/Kaggle/Quora/test.csv")
sample_submission = pd.read_csv("/Users/ganeshshankar/Desktop/work_mac/Personal_Related/Learning/Datascience/Kaggle/Quora/sample_submission.csv")
```

```python
train['question1'] = train['question1'].astype(str)
train['question2'] = train['question2'].astype(str)
```

```python
train['question1'] = train['question1'].str.lower()
train['question2'] = train['question2'].str.lower()
```

```python
def apostrophes(x):
    y = re.sub(r"\b([A-Za-z]+)'re\b", '\\1 are', x)
    z = re.sub(r"\b([A-Za-z]+)'s\b", '\\1 is', y)
    w = re.sub(r"\b([A-Za-z]+)'ve\b", '\\1 have', z)
    u = re.sub(r"\b([A-Za-z]+)'ll\b", '\\1 will', w)
    v = re.sub(r"\b([A-Za-z]+)'n't\b", '\\1 not', u)
    t = re.sub(r"\b([A-Za-z]+)'d\b", '\\1 had', v)
    s = re.sub(r"\b([A-Za-z]+)'m\b", '\\1 am', t)
    return t
```

```python
train['question1'] = train['question1'].apply(apostrophes)
train['question2'] = train['question2'].apply(apostrophes)
```

```python
qtype_list = ['why','when','what','who','where','how','which','should',
              'is','was','are','were','did','do','does','can','could','has','have','will','would','am','any']

def tags(question):
    tokens = word_tokenize(question)
    tokens1 = [token.lower() for token in tokens if type(token) is str]
    return list(set([tag for tag in tokens1 if tag in qtype_list]))
```

```python
def stop_words(s1):
    s1 = str(s1).lower().split()
    stop_words = stopwords.words('english')
    return [w for w in s1 if w not in stop_words]
```

```python
train['question1tag'] = train['question1'].apply(tags)
train['question2tag'] = train['question2'].apply(tags)
train['question1sw'] = train['question1'].apply(stop_words)
train['question2sw'] = train['question2'].apply(stop_words)
```

```python
train.same = train[(train.question1 != train.question2)]
```
